```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import random
import re

url  = 'https://www.cars24.com/buy-used-car/'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html.parser')

Cars_dict = {}
cars_no = 0
no_page = 1

tables = soup.find_all("div", {"class": "col-4"})
print(tables)
```

```python
while cars_no < 4000:
    for table in tables:

        url = 'https://www.cars24.com/buy-used-car/'

        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'html.parser')

        name_value  = table.find('h2', {'class': '_3FpCg'})
        # print(name_value)
        max_power = 110
        if(name_value is None):
          continue
        name = table.find('h2', {'class': '_3FpCg'}).text.split()
        brand = name[1]
        model_date = name[0]
        model_date = int(model_date)
        del name[0]
        del name[0]
        name = " ".join(name)
        model = table.find('p', {'class': 'cvakB'}).text.split()
        transmission = "NA"
        if(len(model)>1):
          transmission = model[-1]
          model.pop()
          model = " ".join(model)
        
        basic_details = table.find('ul', {'class': 'bVR0c'}).text.split()
        # for detail in basic_details:

        kms_driven = basic_details[0]
        kms_driven = kms_driven.replace(',','')
        kms_driven = int(kms_driven)
        owner = basic_details[1][2]
        fuel = basic_details[2][5:-5]
        reg = basic_details[2][-5:13]

        emi = table.find('div', {'class': '_2HFRN'}).text[1:-6]
        emi = emi.replace(',','')
        emi = int(emi)

        price = table.find('div', {'class': '_7udZZ'}).text[1:]
        price = price.replace(',','')

        price = int(price)

        link =table.find('a', {'class': '_9Ue0B'}).get('href')


        more_details = requests.get(link)
        details = more_details.text
        details_soup = BeautifulSoup(details, 'html.parser')
        mileage = random.randint(12,28)
        hp = max_power//(mileage/10)
        overviews = details_soup.find('ul', {"class": "_2KO13"})

        print(name,model_date,brand,transmission,model,kms_driven,owner,fuel,reg,emi,price,mileage,hp)

        cars_no = cars_no + 1
        Cars_dict[cars_no] = [name, model_date, fuel, reg, price, emi, kms_driven, owner,
                              transmission]

cars_dict_df = pd.DataFrame.from_dict(Cars_dict, orient='index',
                                columns=['Name_of_car', 'Model_Date', 'Fuel_Type', 'Location','Price_Rs',
                                      'Emi_Starts_At :','KmS_Driven','Owner',
                                      'Transmission'])
cars_dict_df.to_csv('about_cars24.csv')
```
